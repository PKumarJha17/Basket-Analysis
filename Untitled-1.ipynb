```python
result=list()
```

```python
for i in items:
    s1=basket_input[basket_input[i]>0]
    l1=[]
    for col in s1.columns:
        l1.append(sum(s1[col]))
    result.append(l1)
```

```python
for i in range(len(result)):
    yo=max(result[i])
    for j in range(len(result[i])):
        result[i][j]=result[i][j]/yo
```

```python
df3= pd.DataFrame(result, columns = items) 
```
